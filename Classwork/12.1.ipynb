```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def f(x):
    Z = 24.44321494051954 
    if abs(x) > 7:
        return 0 
    elif abs(x) > 3:
        return 3 * (1 - (x / 7) ** 2) ** 0.5 / Z 
    elif abs(x) > 1:
        return (
            (3 - abs(x)) / 2 -
            3/7 * 10**0.5 * ((3 - x**2 + 2*abs(x))**0.5 - 2)
        )/Z
    elif abs(x) > 0.75:
        return (9 - 8 * abs(x)) / Z 
    elif abs(x) > 0.5:
        return (3 * abs(x) + 0.75) / Z 
    else:
        return 2.25 / Z
    
def g(x):
    return np.random.normal(x,2)

def step(x_t):
    val = g(x_t)
    prob = min(1, f(val) / f(x_t))
    if random.random() < prob:
        return val
    else:
        return x_t
```

```python
start = 1
steps = 500000

val = start
hist = []
for i in range(steps):
    val = step(val)
    if i % 100 == 0 and i !=0:
        hist.append(val)
```

```python
plt.style.use('ggplot')
xs = np.linspace(-7,7,100)
ys = [f(x) for x in xs ]
plt.plot(xs,ys, c = 'r')
plt.hist(hist, bins=50, normed=True, color = 'c')
plt.show()
```
