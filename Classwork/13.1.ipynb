```python
import random
import matplotlib.pyplot as plt

# basic walk
p_h = 0.5
steps = 1000
repeats = 1000

def walk(p,steps):
    place = 0
    array =[place]
    for i in range (steps):
        if random.random() < p:
            place +=1
        else: 
            place -=1
        array.append(place)
    return array

array = walk(p_h, steps)

plt.style.use('ggplot')
plt.plot(array)  
plt.show()

hist = []
for i in range(repeats):
    hist.append(walk(p_h, steps)[-1])
plt.hist(hist)
plt.show()

```
