```python
import random
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
#Random Walk with Drift

steps = list(range(31))

def coin():
    #Returns 1 for Heads, -1 for Tails
    if random.random() < 0.6:
        return 1
    else:
        return -1

end_points = []
plt.style.use('ggplot')
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
for i in range(200):
    path = []
    for step in steps:
        if step == 0:
            path.append(0)
        else:
            path.append(path[-1]+coin())
    plt.plot(steps, path, 'r', alpha = 0.05)
    end_points.append(path[-1])
    plt.plot([31], [path[-1]], marker='o', markersize=3, color="blue", alpha = 0.05)
plt.subplot(1,2,2)
plt.hist(end_points, 10, facecolor='green', alpha=0.75)
plt.show()

```

```python
mean = np.mean(end_points)
var = np.var(end_points)
sd = np.std(end_points)
print (mean)
print (var)
print (sd)
```

```python
#Gambler's Ruin

#do gambler's ruin, but with various upper bounds
# how does is bias the average?
# how does it bias the variance

plt.style.use('ggplot')
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
times = []
for x in range(20):
    r = []
    i = 0
    r.append(100)
    while r[i] >0 and r[i] < 2100:
        i +=1
        w = random.choice([0,1])
        w = 2*w - 1
        r.append(r[-1] + w)
    times.append(len(r))
    plt.plot(r, color = 'r', alpha = 0.3)
plt.subplot(1,2,2)
plt.hist(times, 25, facecolor='green', alpha=0.75)
plt.show()
```

```python
cut_offs = [1000,10000,100000,2000000]

for cut_off in cut_offs:
    plt.style.use('ggplot')
    plt.figure(figsize = (10,5))
    plt.subplot(1,2,1)
    times = []
    for x in range(20):
        r = []
        i = 0
        r.append(100)
        while r[i] >0 and r[i] < 2100 and i <= cut_off:
            i +=1
            w = random.choice([0,1])
            w = 2*w - 1
            r.append(r[-1] + w)
        times.append(len(r))
        plt.plot(r, color = 'r', alpha = 0.3)
        #plt.title = (str(cut_off))
    plt.subplot(1,2,2)
    plt.hist(times, 10, facecolor='green', alpha=0.75)
    plt.show()
    print ("Mean is", np.mean(times) ,"and the Variance is", np.var(times))
```
